5. Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’) Hint. Look for string functions like replace/strip). The data frame will be look like after that:

```python
car_sales['Price'] = car_sales['Price'].str.replace('$', '')
```

```python
car_sales['Price'] = car_sales['Price'].str.replace(',', '')
```

```python
car_sales['Price'] = car_sales['Price'].astype(np.int64)
```

```python
car_sales.info()
```

6. Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
car_sales["Total Sales"]=car_sales['Price'].cumsum()
```

```python
car_sales.info()
```

```python
car_sales.head()
```

7.	Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out “date_range” function and keep the periods=len(car_sales). The data frame will be look like that:

```python
car_sales['Sale Date']=pd.date_range(start=22/2/22, periods=len(car_sales))
```

```python
car_sales.info()
```

```python
car_sales.head()
```

```python
plt.plot(car_sales['Price'],car_sales['Total Sales'])
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs total sales line plot')
```

```python
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer vs Price scatter plot')
```

```python
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer bar plot')
```

```python
plt.hist(car_sales['Make'])
plt.xlabel('Make')
plt.title('Histogram of car make')
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(12,10))

ax1.plot(car_sales['Price'],car_sales['Total Sales'])
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')
ax1.set_title('Price vs total sales line plot')

ax2.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel('Price')
ax2.set_title('Odometer vs Price scatter plot')

ax3.bar(car_sales['Make'],car_sales['Odometer (KM)'])
ax3.set_xlabel('Make')
ax3.set_ylabel('Odometer (KM)')
ax3.set_title('Make vs Odometer bar plot')

ax4.hist(car_sales['Make'])
ax4.set_xlabel('Make')
ax4.set_title('Histogram of car make')
```

```python
sns.boxplot(car_sales['Colour'],car_sales['Total Sales'])
```

```python
sns.lmplot(x='Total Sales',y='Odometer (KM)',
           data=car_sales,ci=False, hue='Make')
```

```python
heat_map=car_sales[['Total Sales','Price','Odometer (KM)']].corr(method='pearson')
cols=['Total Sales','Price','Odometer (KM)']
axis=sns.heatmap(heat_map,annot=True,xticklabels=cols,yticklabels=cols)
```

There is a negative correlation between how many miles are on a car
and Total Sales
